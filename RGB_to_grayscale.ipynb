```python
from PIL import Image
import os
```

```python
RGB_folder = "./data/"
grayscale_folder = "./data_grayscale/"
```

```python
for filename in os.listdir(RGB_folder):
    image = Image.open(os.path.join(RGB_folder, filename))
    if image.mode != "RGB":
        image.close()
        print(f"{filename} is not an RGB image")
        os.remove(os.path.join(RGB_folder, filename))
```

```python
if not os.path.exists(grayscale_folder):
    os.makedirs(grayscale_folder)
```

```python
for file_name in os.listdir(RGB_folder):
    img = Image.open(os.path.join(RGB_folder, file_name))
    gray = img.convert("L")
    gray.save(os.path.join(grayscale_folder, file_name))    
print("Conversion to grayscale is done")
```

```python
for filename in os.listdir(grayscale_folder):
    img = Image.open(os.path.join(grayscale_folder, filename))
    if img.mode != "L":
        img.close()
        print(f"{filename} is not a grayscale image")
        os.remove(os.path.join(grayscale_folder, filename))
```
